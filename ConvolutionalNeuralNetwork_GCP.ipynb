```python
#PART-3
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense

# Load the Iris dataset

iris_data = pd.read_csv("/content/Iris.csv")

# Extract the feature matrix and target variable

array = iris_data.values
X = array[:,1:5]
Y = array[:,5]
test_size = 0.33
random_seed = 7
#training and testing split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size, random_state=random_seed)

# Encode the target variable as integers
encoder = LabelEncoder()
Y_train_encoded = encoder.fit_transform(Y_train)
Y_test_encoded = encoder.transform(Y_test)

# Convert the integers to one-hot encoded vectors
Y_train_onehot = np_utils.to_categorical(Y_train_encoded)
Y_test_onehot = np_utils.to_categorical(Y_test_encoded)

# Define the model architecture
def baseline_model():
    model = Sequential()
    model.add(Dense(6, input_dim=4, activation='relu'))
    model.add(Dense(6, activation='relu'))
    model.add(Dense(3, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

```

```python
# Compile the model
model = baseline_model()

# Train the model on the training set
model.fit(X_train.astype('float32'), Y_train_onehot.astype('float32'), epochs=150, batch_size=10)
```

```python
from pandas import Series, DataFrame
from sklearn.preprocessing import LabelEncoder

from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold


#evaluating model
evaluator = KerasClassifier(build_fn=baseline_model, epochs=200, batch_size=5)
kfold = KFold(n_splits=10, shuffle=True, random_state=random_seed)

# Cast the test data to float32
X_test = X_test.astype('float32')
onehot_Y_test = Y_test_onehot.astype('float32')

results = cross_val_score(evaluator, X_test, onehot_Y_test, cv=kfold)

print("Baseline: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))

# Evaluate the model on the test set
scores = model.evaluate(X_test, onehot_Y_test)

# Print the model's accuracy on the test set
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
